**Question 4 : Select columns ‘Y2017’ and ‘Area’, Perform a groupby operation on ‘Area’.  Which of these Areas had the highest sum in 2017?**

```python
#Select Data
data_2017 = data[['Y2017', 'Area']]
data_2017.head()
```

```python
grouped17 = data_2017.groupby(by = ['Area']).sum()  #group data
grouped17.sort_values(by = 'Y2017', ascending = False) #sort values
```

**Question 6: Perform a groupby operation on ‘Element’.  What year has the highest sum of Stock Variation?**

```python
#group by Element
grpby_Element = data.groupby(by = ['Element']).sum()
grpby_Element
```

```python
grpby_Element_years = grpby_Element[['Y2014', 'Y2015', 'Y2016','Y2017', 'Y2018']] # take the years
grpby_Element_years.loc['Stock Variation'].idxmax()  #find the year
```

**Question 8: Which year had the least correlation with ‘Element Code’?**

```python
data_corr = data.corr()[['Y2014', 'Y2015', 'Y2016','Y2017', 'Y2018']] #get correlation with years
data_corr.loc['Element Code'].idxmin()  ## year with least correlation
```

**Question 14: What is the total number and percentage of missing data in 2014 to 3 decimal places?**

```python
#missing data 2014
missing_vals = uncleaned_data['Y2014'].isnull().sum()  #missing values
percentage = uncleaned_data['Y2014'].isnull().sum() / uncleaned_data['Y2014'].shape[0]  #fraction

print('The number of missing values are : {} and the % is : {} .'.format(missing_vals, percentage*100))
```
